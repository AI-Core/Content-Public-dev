## Security groups
--operates at the instance level
--stateful. so any accepted inbound request will also be allowed outbound automatically
--can only be customised to allow
--You can specify the source/destination, port range, and protocol for each inbound/outbound rule
--can use multiple SGs on the same resource(s)

## NACLs
--like a firewall which controls traffic to and from subnets
--stateless. inbound and outbound rules evaluated exclusively
--one nacl per subnet. new subnets get a default nacl. each subnet must be associated with a nacl
    --default nacl accepts everything inbound/outbound

--newly created nacls will deny everything by default
--can be customised to allow or deny
--nacl rules. 
    --rules are numbered, lower numbers get a higher precedence
--gotta remember to think about ephemeral ports when setting up nacls and their port ranges:
    --for any two endpoints to establish a connection they must use ports
    --clients connect to a *defined port* on the target, and expect a response on an *ephemeral port*

## firewalls
--what is a firewall? a network security device that monitors incoming and outgoing network traffic and decides whether to allow or block specific traffic based on a defined set of security rules.

--aws network firewall
    --intrusion detection and prevention service for Amazon Virtual Private Cloud (Amazon VPC). You can combine Network Firewall with services and components that you use with your VPC, for example an internet gateway, a NAT gateway, a VPN, or a transit gateway
    --made up of the following components:

        --rule group - a collection of criteria which decides whether packets in a traffic flow are allowed to pass or drop. rule groups are either stateless or stateful: 
            --Stateless rule groups evaluate packets in isolation, while stateful rule groups evaluate them in the context of their traffic flow. 
        --firewall policy - a reusable set of stateless and stateful rule groups, along with some policy-level behaviour such as what happens if a stream of traffic breaks midstream, default actions for both stateless and stateful traffic
        --Firewall â€“ Connects the inspection rules in the firewall policy to the VPC that the rules protect. Each firewall requires at least one firewall policy. The firewall additionally defines settings like how to log information about your network traffic.

--WAF:
    --AWS WAF is a web application firewall that lets you monitor the HTTP(S) requests that are forwarded to your protected web application resources.
    --AWS WAF lets you control access to your content. Based on criteria that you specify, such as the IP addresses that requests originate from
    --this is done by defining a web ACL then associating it with one or more web application resources, which pass all incoming traffic through the WAF first for inspection
        --the ACL is made up of rules which define traffic patterns to look for in requests and to specify the actions to take on matching requests. the actions can be:
            --Allow the requests to go to the protected resource for processing and response.
            --Block the requests.
            --Count the requests.
            --Run CAPTCHA or challenge checks against requests to verify human users and standard browser use.
    
    --WAF components:
            --web ACLs, which are made up of rule groups, which define criteria for inspecting web requests and then determine the action to take
                --you can set a default action for the web ACL that decides whether to block or allow any requests that the rules haven't already dealt with.
            --rules, which contain one or more statements that define the inspection criteria, then the action to take if a request meets the criteria. E.G:
                --defining IP addresses or ranges which requests come from
                --country or location which requests originate from
                --you can check for specific known malicious request attributes such as the query string, the request header value or body values
            --rule group. reusable groups of rules. you can use predefined rule groups on AWS Marketplace


## What is Network Monitoring?
--Network monitoring or logging is the process of documenting every event that takes place on a device. 

Network engineers and administrators typically use network monitoring tools to collect the following types of metrics from network devices:

Uptime
The amount of time that a network device successfully sends and receives data.

CPU utilization/also called 'compute'?
The extent to which a network device has used its computational capacity to process input, store data, and create output.

Bandwidth usage
The amount of data, in bytes, that is currently being sent or received by a specific network interface. Engineers track both the volume of traffic being sent, and the percentage of total bandwidth that is being utilized.

Throughput
The rate of traffic, in bytes per second, passing through an interface on a device during a specific time period. Engineers typically track throughput of a single interface, and the sum of the throughput of all interfaces on a single device.

Interface errors/discards
These are errors on the receiving device that cause a network interface to drop a data packet. Interface errors and discards can stem from configuration errors, bandwidth issues, or other reasons.

IP metrics
IP metrics, such as time delay and hop count, can measure the speed and efficiency of connections between devices.

Note that in cloud environments, companies purchase compute and network resources from cloud vendors who maintain the physical infrastructure that will run their services or applications. Cloud hosting therefore shifts the responsibility of managing the physical hardware onto the cloud vendor.

## cloudwatch
--aws service which monitors your aws resources and applications
--you can collect and track metrics about each service you use, as well as create custom dashboard displaying the metrics
--you can create alarms to watch metrics and send automatic notifications or automatically change resources when a threshold is breached. 
    --e.g. you can monitor CPU usage of an EC2 instance then use that data to decide whether to launch a new instance to handle increased load.

## cloudtrail
--event tracking system for an AWS account
--Actions taken by a user, role, or an AWS service are recorded as events in CloudTrail.
--Events include actions taken in the AWS Management Console, AWS Command Line Interface, and AWS SDKs and APIs.
--cloudtrail provides three ways to record events:
    --event history: searchable record of the past 90 days of events in an AWS region. no charges for using this. you automatically have access to Cloudtrail Event History by default when you create an AWS account.
    --cloutrail lake: a managed data lake for capturing, storing, analysing user and API activity. you can keep the data for up to 10 years or up to 7 years, depending on which pricing plan you choose. you can also import existing cloudtrail logs from S3 buckets into Cloudtrail Lake to use its dashboarding features.
    --Trails: captures and records events, storing them in an S3 bucket, with optional delivery to cloudwatch logs or other external security and monitoring solutions.
